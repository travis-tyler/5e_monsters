# Clean

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as st
```

```python
# Read in data
df = pd.DataFrame(pd.read_csv('../resources/dnd_monsters.csv'))
df.head()
```

```python
# Remove unwanted columns
df = df.drop(columns=['url','type','speed','align','source'])
df.head()
```

```python
# Convert legendary status to boolean
df['legendary'] = df['legendary'].replace({np.nan:0,'Legendary':1})

# Convert size descriptor to int values
df['size'] = df['size'].replace({'Tiny':1,'Small':2,'Medium':3,'Large':4,'Huge':5,'Gargantuan':6})

# Change CR fractions to float
df['cr'] = df['cr'].replace({'1/8':'.125','1/4':'.25','1/2':'.5'}).astype('float64')

# Drop columns with any null values
df = df.dropna(how='any')

df
```

# Explore

# Features and ML training

```python
from sklearn.pipeline import make_pipeline
from sklearn.linear_model import LinearRegression, ElasticNet, Ridge, Lasso
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer, make_column_transformer, make_column_selector
from sklearn.metrics import mean_squared_error
```

```python
X = df.drop(columns=['name','cr'])
y = df[["cr"]]
print(X.shape, y.shape)
```

```python
ct = make_column_transformer(
    (StandardScaler(), make_column_selector(dtype_include=np.number)),
    (OneHotEncoder(), make_column_selector(dtype_include=object))
)
ct.fit_transform(X)
```

```python
# Split the data into training and testing
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
models = [LinearRegression(), ElasticNet(alpha=.01), Ridge(alpha=.01), Lasso(alpha=.01)]

for model in models:
    lr_model = make_pipeline(ct, model)

    #other models: LinearRegression(), ElasticNet(alpha=.01), Ridge(alpha=.01), Lasso(alpha=.01)

    lr_model.fit(X_train, y_train)

    training_score = lr_model.score(X_train, y_train)
    testing_score = lr_model.score(X_test, y_test) 
    MSE = mean_squared_error(y_test, lr_model.predict(X_test))

    print(f'{model}:')
    print(f'Training R2 Score: {training_score}')
    print(f'Testing R2 Score: {testing_score}')
    print(f'MSE: {MSE}')
    print('\n------\n')
```
