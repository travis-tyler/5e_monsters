# Clean

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as st
```

```python
# Read in data
df = pd.DataFrame(pd.read_csv('../resources/dnd_monsters.csv'))
df.head()
```

```python
# Remove unwanted columns
df = df.drop(columns=['url','type','speed','align','source'])
df.head()
```

```python
# Convert legendary status to boolean
df['legendary'] = df['legendary'].replace({np.nan:0,'Legendary':1})

# Convert size descriptor to int values
df['size'] = df['size'].replace({'Tiny':1,'Small':2,'Medium':3,'Large':4,'Huge':5,'Gargantuan':6})

# Change CR fractions to float
df['cr'] = df['cr'].replace({'1/8':'.125','1/4':'.25','1/2':'.5'}).astype('float64')

# Drop columns with any null values
df = df.dropna(how='any')

df
```
